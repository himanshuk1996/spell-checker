```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
stopwords=stopwords.words('english')
```

```python
import string
import collections
```

```python
data = pd.read_excel('ticketdata.xlsx')
data['corpus'] = data['reason'] +" "+ data['initial_description']
```

```python
def remove_stopwords(sentence, stopwords):
	sentencewords = sentence.split()
	resultwords  = [word for word in sentencewords if word.lower() not in stopwords]
	result = ' '.join(resultwords)
	return result
```

```python
REPLACE_NO_SPACE = re.compile("[^A-Za-z\s]")
stemmer = nltk.SnowballStemmer('english')
lemmatizer=nltk.WordNetLemmatizer()
def preprocess_reviews(data):
    data = [REPLACE_NO_SPACE.sub("", str(line).lower()) for line in data]
    data = [remove_stopwords(w,stopwords) for w in data]
    #data = [lemmatizer.lemmatize(w) for w in data]
    #data = [stemmer.stem(w) for w in data]
    return data
```

```python
raw_input=preprocess_reviews(data['corpus'])
```

```python
raw_input
```

```python
my_words=[]
for line in raw_input:
    line=line.split(' ')
    for word in line:
        my_words.append(word)
```

```python
filtered_words=[word for word in my_words if len(word)<13]
```

```python
my_vocabulary=list(set(filtered_words))
```

```python
keypad='qwertyuioplkjhgfdsazxcvbnm'
```

```python
reverse_dictionary={}
j=0
for letter in keypad:
    reverse_dictionary[j+1]=letter
    j=j+1    
```

```python
dictionary=dict(zip(reverse_dictionary.values(), reverse_dictionary.keys()))
```

```python
counter=collections.Counter(filtered_words)
```

```python
counter.most_common(3000)
```

```python
counter.get('hal')
```

```python
better_vocabulary=[]
for word in my_vocabulary:
    if(counter.get(word)>1):
        better_vocabulary.append(word)
```

```python
def vectorizer(list_of_words):
    vects=[]
    for word in list_of_words:
        vector=[0]*12
        j=0
        for alphabet in word:
            vector[j]=dictionary[alphabet]
            j=j+1
        vector[0]=vector[0]*10
        vects.append(vector)
    return vects
```

```python
vects=vectorizer(better_vocabulary)
```

```python
from sklearn.neighbors import NearestNeighbors
kn = NearestNeighbors(n_neighbors=5, radius=1.0, algorithm='auto', 
                      leaf_size=30, metric='minkowski', p=2, metric_params=None, n_jobs=-1)
```

```python
kn.fit(vects)
```

```python
text=vectorizer(['helo'])
t=kn.kneighbors(text,5,return_distance=True)
```
